# How to run CASA PlotMS in a notebook?

You will need to install `casatasks`, `casaplotms`, `casadata`, and `pyvirtualdisplay` to run this notebook.

## Setup the virtual frame buffer

```python
from pyvirtualdisplay import Display
display = Display(visible=0,size=(2048,2048))
display.start( )
```

## Plot the visibility amplitude

```python
from casaplotms import plotms

ms_name = "hi_from_heald.ms"

# Plot baseline 10&11 
plotms(
	vis=ms_name,
	xaxis="freq",
	yaxis="amp",
	field="0",
    antenna="10&11",
	averagedata=True, avgtime="1e10",
	ydatacolumn='data',
	plotfile='hi_0.png',
	showgui=False,highres=True,overwrite=True
)
```

The above writes the plot to disk. Display the produced image. 

```python
from IPython.display import Image
Image(filename="hi_0.png")
```
